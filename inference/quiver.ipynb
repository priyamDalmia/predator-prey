```python
import numpy as np
import matplotlib.pyplot as plt
data_i = np.load("../data_test.npy", allow_pickle=True)
```

```python
mat_X = data_i[1,:,:] - data_i[0,:,:]
mat_Y = data_i[2,:,:] - data_i[3,:,:]
mat_X = np.array(mat_X, dtype=np.float32)
mat_Y = np.array(mat_Y, dtype=np.float32)
color = np.array(data_i[-1,:,:], dtype=np.float32)
```

```python
plot_X, plot_Y = np.meshgrid(range(9), range(9))
plt.quiver(plot_Y, -plot_X, -mat_X, -mat_Y, color)
plt.show()
```

```python
from matplotlib.ticker import AutoMinorLocator
plot_X, plot_Y = np.meshgrid(range(9), range(9))
ax = plt.quiver(plot_X, -plot_Y, -mat_X, -mat_Y, color, pivot='tip')
plt.minorticks_on()
plt.grid(which='minor', linestyle=':', linewidth='0.5', color='black')
plt.show()
```

```python
from matplotlib.ticker import AutoMinorLocator
plot_X, plot_Y = np.meshgrid(range(9), range(9))
plt.quiver(plot_Y, -plot_X, -mat_X, -mat_Y, color, pivot='mid', units='width')
ax = plt.gca()
#ax.set_xticks([x-0.5 for x in range(1,9)],minor=True)
#ax.set_yticks([-x-0.5 for x in range(0,9)],minor=True)
plt.grid(which="minor", ls="--",lw=1, alpha=0.5)
plt.text(2, -2, "D0")
plt.show()
```
